##                                        Data Visualisation

## 1) What is the average total amount spent by each customer type and gender

## 2) Which product line has highest average rating 


```python
#Grouping product line by rating
grouped_product=df.groupby('Product line')['Rating'].mean().reset_index()
grouped_product=grouped_product.round(1)
grouped_product_sort=grouped_product.sort_values(by='Rating',ascending=False)
hori_bar=sns.barplot(y='Product line',x='Rating',data=grouped_product_sort,palette='Blues_r')
hori_bar.bar_label(hori_bar.containers[0])
# palette.reverse()
```

## 6) Which product line has the highest average sales?

```python
#Grouping Product line by rat
grouped_productline_total=df.groupby('Product line')['Total'].mean().reset_index(drop=False)
#Rounding Rating column to 1 number
grouped_productline_total=grouped_productline_total.round(1)
annotations=grouped_productline_total['Total']
for i,label in enumerate(annotations):
    plt.annotate(label,(grouped_productline_total['Product line'][i],grouped_productline_total['Total'][i]))
    plt.xticks(fontsize=6)
#Making a scatter plot for Product lines` average sales
sns.scatterplot(x='Product line',y='Total',data=grouped_productline_total)
plt.axhline(y=np.nanmean(df.Total), color='red', linestyle='--', linewidth=3, label='Avg')
plt.legend()
#setting the title
plt.title("Average Sales for Product Lines")
```

## 10) What are the interests of each gender

```python
cross_tab=pd.crosstab(index=df['Gender'],columns=df['Product line'])
cross_tab_prop=pd.crosstab(index=df['Gender'],columns=df['Product line'],normalize='index')
```

```python
cross_tab_prop.plot(kind='barh',stacked=True,colormap='coolwarm')
plt.legend(bbox_to_anchor=(0.5, 1.02),loc="lower center",borderaxespad=0,frameon=False,ncol=3,  prop = {'size' : 8})
plt.xlabel('Propotion')
for n, x in enumerate([*cross_tab.index.values]):
    for (proportion, count, y_loc) in zip(cross_tab_prop.loc[x],
                                          cross_tab.loc[x],
                                          cross_tab_prop.loc[x].cumsum()):
                
        plt.text(x=(y_loc - proportion) + (proportion / 2)-0.05,
                 y=n - 0.12,
                 s=f'{count}\n({np.round(proportion * 100, 1)}%)', 
                 color="black",
                 fontsize=10,
                 fontweight="bold")
```

* From the last stacked bar chart , we can see that Females are interested in Fashion accessories , they aren`t interested in    Health and Beauty Products
* Males are interested in all categories , but they are interested in Health and Beauty the most

## 12) Which Product line has had the highest average sales city-wise

```python
grouped_productline=df.groupby(['City','Product line'])['Total'].mean().reset_index()
#visualize using bar plot
ax=sns.barplot(x='Product line',y='Total',hue='City',data=grouped_productline,palette='crest')
plt.legend(bbox_to_anchor=(0.3, 1.00),loc="upper center",borderaxespad=0,frameon=False,ncol=3,  prop = {'size' : 6})

plt.xticks(rotation=80)
#Data labels
for label in ax.containers:
    ax.bar_label(label,fontsize=6,weight='bold')
#Set title
plt.title('Total Sales Per Product lines and City')
```

* In city Yangon, customers spend on Health and Beauty the least and Home and Lifestyle the most.
* In city Mandalay, customers spend on Food and Beverages the least and Sports and travel the most.
* In city Naypyidaw, customers spend on Home and lifestyle the least and Food and Beverages the most
