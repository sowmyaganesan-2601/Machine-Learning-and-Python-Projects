# Missing Value Treatment 

since % of these records compared to total dataset is very low i.e. 0.15% ,it is safe to ignore them for further processing

```python
df.dropna(how='any',inplace=True)
```

Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
print(df['tenure'].max())
```

```python
df_data=df.copy()
```

```python
#Group the tenure in bins of 12 months
labels=["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
```

```python
df_data['tenure_group']=pd.cut(df.tenure,range(1,80,12) ,right=False,labels=labels)
```

```python
df_data['tenure_group'].value_counts()
```

 ## Data Exploration

1. Plot distibution of individual predictors by churn



Univariate Analysis

```python
for i,predictor in enumerate(df_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
         plt.figure(i)
         sns.countplot(data=df_data,x=predictor,hue='Churn')
```

```python
df_data.head()
```

2. Convert the target variable 'Churn' in a binary numeric variable i.e. Yes=1 ; No = 0



```python
df_data['Churn']=np.where(df_data.Churn=='Yes',1,0)
df_data.head(5)
```

3. Relationship between Monthly Charges and Total Charges



```python
sns.lmplot(data=df_data, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Total Charges increase as Monthly Charges increase - as expected.



Churn by Monthly Charges and Total Charges

```python
Mth = sns.kdeplot(df_data.MonthlyCharges[(df_data["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(df_data.MonthlyCharges[(df_data["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

Churn is high when monthly charges are high

```python
df_dummies=pd.get_dummies(df_data)
df_dummies
```

```python
# create a correlation matrix
corr =df_dummies.corr()

# select only the column with the 'Churn' variable
corr_churn = corr['Churn']

# plotting
sns.heatmap(corr_churn, cmap=sns.color_palette("blend:#7AB,#EDA", as_cmap=True), annot=True)

# show the plot
plt.show()
```
